Problem 1 — Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Recharge must be at least ₹50")
        elif amount not in valid_plans:
            print("Invalid plan. Available plans:", valid_plans)
        else:
            print("Recharge Successful for plan ₹", amount)
            break

validate_recharge()
```

Problem 2 — Inventory Reorder Alert System

```python
def check_inventory(stock):
    for product in stock:
        if stock[product] < 15:
            print(product, "→ Reorder Alert")
        else:
            print(product, "→ Stock OK")


inventory = {
    "Rice": 20,
    "Wheat": 10,
    "Sugar": 25,
    "Oil": 12
}

check_inventory(inventory)
```

Problem 3 — Student Result Processing System

```python
def student_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")


marks_list = [60, 55, 70, 40]
student_result(marks_list)
```

Problem 4 — Cab Fare Estimator with Retry Option

```python
def calculate_fare(km, peak):
    fare = 50 + (km * 12)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:
    km = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(km, peak)

    print("Total Fare: ₹", round(total_fare, 2))

    retry = input("Calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

Problem 5 — Employee Attendance Eligibility Checker

```python
def check_attendance(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_list = ["P", "P", "A", "P", "P", "A", "P"]
result = check_attendance(attendance_list)
print("Eligibility:", result)
```

Problem 6 — Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        print("Password too short")
        return

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        print("Strong Password")
    else:
        print("Weak Password")


user_password = input("Enter password: ")
check_password(user_password)
```
