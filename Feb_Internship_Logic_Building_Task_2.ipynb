Problem 1: Unique Words in a Sentence
Problem Statement
Given a sentence, identify all unique words and display:
1. The count of unique words
2. The unique words themselves

```python
# Sentence given
sentence = "python is easy and python is powerful"

# Split sentence into words
words = sentence.split()

# Remove duplicates using set
unique_words = set(words)

# Print result
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data
Problem Statement
Store employee names and their salaries, and determine which employee has the highest
salary.

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_name = ""
highest_salary = 0

# Finding highest salary
for name in employees:
    if employees[name] > highest_salary:
        highest_salary = employees[name]
        highest_name = name

print("Highest Salary:", highest_name, "-", highest_salary)
```

Problem 3: Find Maximum and Minimum Values
Problem Statement
Find the maximum and minimum values from a list of numbers.

```python
numbers = [45, 22, 89, 10, 66]

max_value = numbers[0]
min_value = numbers[0]

# Loop through list
for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

Problem 4: Count Products Above a Price Threshold
Problem Statement
Count how many products have a price greater than 1000

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count = count + 1

print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage
Problem Statement
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

```python
attendance = ["P", "P", "A", "P", "P"]

present = 0

for day in attendance:
    if day == "P":
        present = present + 1

percentage = (present / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

Problem 7: Count Character Frequency
Problem Statement
Count how many times each character appears in a given string.

```python
text = "python"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] = frequency[char] + 1
    else:
        frequency[char] = 1

print(frequency)
```

Problem 10: Calculate Average Marks
Problem Statement
Calculate the average marks of a student from a list of marks.

```python
marks = [70, 80, 65, 75]

total = 0

for m in marks:
    total = total + m

average = total / len(marks)

print("Average Marks:", average)
```
