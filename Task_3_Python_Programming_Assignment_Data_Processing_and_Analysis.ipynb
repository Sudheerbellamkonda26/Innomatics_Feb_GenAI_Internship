Problem 1 — Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

top_performers = []

# Find employees who have highest score
for name in employees:
    if employees[name] == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers),
      "(Score:", highest_score, ")")
```

Problem 2 — Search Query Keyword Analysis

```python
text = "Buy mobile phone buy phone online"

# Convert to lowercase
text = text.lower()

# Remove punctuation (not needed here but safe)
import string
for p in string.punctuation:
    text = text.replace(p, "")

words = text.split()

frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Print only repeated keywords
result = {}
for word in frequency:
    if frequency[word] > 1:
        result[word] = frequency[word]

print(result)
```

Problem 3 — Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index in range(len(sensor_readings)):
    value = sensor_readings[index]

    if value % 2 == 0:   # check even number
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem 4 — Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_users = len(emails)

# Calculate percentage
for domain in domain_count:
    percentage = (domain_count[domain] / total_users) * 100
    print(domain + ":", str(int(percentage)) + "%")
```

Problem 5 — Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
average = sum(sales) / len(sales)

# Check spike (>30% above average)
threshold = average * 1.3

for i in range(len(sales)):
    if sales[i] > threshold:
        print("Day", i + 1, ":", sales[i])
```

Problem 6 — Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

frequency = {}

for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1

# Print duplicates only
for user in frequency:
    if frequency[user] > 1:
        print(user, "→", frequency[user], "times")
```
