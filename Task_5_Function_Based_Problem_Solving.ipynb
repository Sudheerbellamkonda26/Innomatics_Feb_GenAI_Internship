Problem 1 — Post Engagement Analyzer

```python
def analyze_post(likes):
    total_likes = 0

    for l in likes:
        total_likes += l

    print("Total Likes:", total_likes)

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


likes_list = [200, 300, 150, 400]
analyze_post(likes_list)
```

Problem 3 — Rainfall Adequacy Checker

```python
def check_rainfall(rainfall, required):
    total = 0

    for r in rainfall:
        total += r

    average = total / len(rainfall)
    print("Average Rainfall:", average)

    if average >= required:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


rain_data = [70, 75, 68, 80, 67]
check_rainfall(rain_data, 70)
```

Problem 4 — Duplicate Account Detection

```python
def check_duplicates(usernames):
    unique = set(usernames)

    if len(unique) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


users = ["user1", "user2", "user3", "user1"]
check_duplicates(users)
```

Problem 6 — Premium Crop Price Filter

```python
def premium_crops(prices):
    premium = []

    for p in prices:
        if p > 2000:
            premium.append(p)

    print("Premium Crops:", premium)


crop_prices = [1500, 2500, 1800, 3200]
premium_crops(crop_prices)
```

Problem 7 — Application Health Checker

```python
def check_system(errors):
    print("Error Count:", errors)

    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


check_system(7)
```

Problem 9 — Student Attendance Eligibility

```python
def check_exam_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


attendance_list = ["P", "P", "A", "P", "P"]
check_exam_eligibility(attendance_list)
```

Problem 10 — Smart Electricity Bill Analyzer

```python
def calculate_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    print("Total Bill:", bill)

    if bill < 500:
        print("Usage Level: Low Usage")
    elif bill <= 1500:
        print("Usage Level: Moderate Usage")
    else:
        print("Usage Level: High Usage")


calculate_bill(250)
```
